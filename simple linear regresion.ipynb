```python
data = data.rename(columns={'Delivery Time': 'dt', 'Sorting Time': 'st'})
data
```

```python
import statsmodels.formula.api as smf
model=smf.ols("dt~st",data=data).fit() 
model.params 
```

```python
model.summary() 
```

```python
xlog=np.log(data["st"]) 
model1=smf.ols("dt~xlog",data=data).fit()
model1.summary()
```

```python
ylog=np.log(data["dt"])
model2=smf.ols("ylog~st",data=data).fit() 
model2.summary()
```

```python
model13=smf.ols("ylog~xlog",data=data).fit()
model13.summary()
```

```python
xsqrt=np.sqrt(data["st"])
model4=smf.ols("dt~xsqrt",data=data).fit() 
model4.summary() 
```

```python
ysqrt=np.sqrt(data["dt"])
model5=smf.ols("dt~xsqrt",data=data).fit() 
model5.summary()
```

```python
model6=smf.ols("ysqrt~xsqrt",data=data).fit() 
model6.summary() 
```
